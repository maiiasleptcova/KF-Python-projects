## Splitting the company and indicator

```python
# splitting Name
df[['Company Name','Indicator']] = df['Name'].str.split(' - ',expand=True, n=1)
```

```python
#df[df['Company Name'].str.contains('SUSP',case=True)]
```

```python
# delete original columns
del df['Name']
del df['Code']
```

```python
df['Indicator Code'].unique()
```

```python
# fix Code names
df['Indicator Code'] = df['Indicator Code'].str.replace(')', '')
```

```python
df['Indicator Code'].unique()
```

```python
#df[df['Indicator'].str.contains('SUSP',case=True)]
```

```python
# some of the variable names contain unnecessary additions like, SUSPENDED or DELISTED
# I want to delete these, they are at the beginning of the string
# I want to take X characters from the right side of the string
# X is the total length of string minus the length of unnecessary characters
# X is slice_pos
df['slice_pos']= df['Indicator'].astype(str).map(len)-len('SUSP.01/07/19 - ')
df['slice_pos2']= df['Indicator'].astype(str).map(len)-len('DELIST.30/07/21 - ')
```

```python
#df[df['Indicator'].str.contains('SUSP.',case=True)]
```

```python
# now do the "taking from the right", for strings containing extra characters
df['Indicator'] = np.where(df['Indicator'].str.contains('SUSP.',case=True), df.apply(lambda r: r['Indicator'][-r['slice_pos']:], axis=1)
                      ,df['Indicator'])
df['Indicator'] = np.where(df['Indicator'].str.contains('DELIST.',case=True), df.apply(lambda r: r['Indicator'][-r['slice_pos2']:], axis=1)
                      ,df['Indicator'])
```

```python
# remove SUSP from company name
df['Company Name'] = df['Company Name'].str.replace(' SUSP', '')
```

```python
# remove slice position vars
del df['slice_pos']
del df['slice_pos2']
```

## False/True columns clean-up

```python
#set(esg['Agrochemical 5 % Revenue'].unique()).issubset({'0', 'False'}) 
```

```python
# set the company codes and years aside
esg_year_comp = esg.iloc[:,:2]
```

```python
esg_year_comp.head()
```

```python
# the vars are here 
esg_vars = esg.iloc[:,2:]
```

```python
esg_vars.head()
```

```python
# for columns that have only True, False, zero and NaN values, replace zeros with NaNs
# convert the rest of the columns to numeric, turning True or False values to NaNs
# the distinctive difference that allows to do the right conversion is that
# the only other non-NaN value that True/False columns have is zero
# the numeric columns are mostly numeric, with a small amount of True/False values
# therefore the if/else condition is about being a subset of {'0', 'False', 'True', np.nan} set
# having "issubset" instead of "==" allows to account for columns with any combination of set elements, e.g {'0', 'False'} or 
# {'0', 'True', np.nan} etc

for column in esg_vars:
    # these are True/False columns
    if set(esg_vars[column].unique()).issubset({'0', 'False', 'True', np.nan})  :
        esg_vars[column] = esg_vars[column].replace("0", np.nan)
    else:
        # These are numeric columns
        esg_vars[column] = esg_vars[column].apply(pd.to_numeric, errors='coerce')  
```

```python
esg_vars.shape
```

```python
# if the column True/False outcome doesn't vary (only True or only False values), drop the column
for column in esg_vars:
    if (set(esg_vars[column].unique()) == {'True', np.nan}) | (set(esg_vars[column].unique()) == {'False', np.nan})  :
        del esg_vars[column]
    else:
        continue  
```

```python
esg_vars.shape
```

```python
# now combine back together
esg = pd.concat([esg_year_comp, esg_vars], axis=1)
```
